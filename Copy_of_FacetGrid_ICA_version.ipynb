```python
#Please read titanic.csv from the DATA folder
titanic = pd.read_csv('/content/drive/MyDrive/DATA/titanic.csv')

#Change 1s to "Survived" under the Survived column
titanic["Survived"] = titanic["Survived"].replace(1, "Survived")

#Get rows which don't have "?" under Pclass column and move to the next step.
filtered_titanic = titanic[titanic["Pclass"] != "?"]

print(filtered_titanic.head())
```

Create a multipanel plot which draws density plots for each class-survived combination. Please also use the hue parameter based on the Sex attribute to have two density plots in each panel with different colors.

```python
#you can uncomment below when you come up with your density plots for each class-survived combination
#titles = ['Died in Class 3','Died in Class 1', 'Died in Class 2', 'Survived in Class 3','Survived in Class 1', 'Survived in Class 2']
#for ax,title in zip(g.axes.flatten(),titles):
#    ax.set_title(title )


# Step 2: Change 1s to "Survived" under the Survived column
titanic["Survived"] = titanic["Survived"].replace(1, "Survived")
titanic["Survived"] = titanic["Survived"].replace(0, "Died")

# Step 3: Get rows which don't have "?" under Pclass column
filtered_titanic = titanic[titanic["Pclass"] != "?"]

# Ensure Pclass column is treated as a categorical variable
filtered_titanic["Pclass"] = filtered_titanic["Pclass"].astype(int)

# Set up the grid for the plots
g = sns.FacetGrid(filtered_titanic, col="Pclass", row="Survived", hue="Sex", margin_titles=True, aspect=1.2)
g.map(sns.kdeplot, "Age", fill=True).add_legend()

# Titles for the plots
titles = ['Died in Class 3', 'Died in Class 1', 'Died in Class 2', 'Survived in Class 3', 'Survived in Class 1', 'Survived in Class 2']
for ax, title in zip(g.axes.flatten(), titles):
    ax.set_title(title)

plt.show()
```
